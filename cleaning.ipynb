```python
from utils.loader import load_csv, save_csv
import os
```

```python
data_dir = "./data"
folktales_csv = "folk_tales_deduplicated.csv"

raw_dir = os.path.join(data_dir, "raw")
raw_folktales = os.path.join(raw_dir, folktales_csv)

processed_dir = os.path.join(data_dir, "processed")
processed_folktales = os.path.join(processed_dir, folktales_csv)
```

```python
df = load_csv(raw_folktales)
df.describe()
```

```python
df_clean = df.dropna(how="any")

print("Original rows:", len(df))
print("After drop:", len(df_clean))
print("Dropped rows:", len(df) - len(df_clean))
```

```python
save_csv(processed_folktales, df_clean)
```
